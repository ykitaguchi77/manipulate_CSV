#**Create dataframe with Pandas**

#**リスト形式でデータフレームを作成**

```python
import pandas as pd
 
# initialize data of lists.
data = {'Name':['Tom', 'nick', 'krish', 'jack'],
        'Age':[20, 21, 19, 18]}
 
# Create DataFrame
df = pd.DataFrame(data)
 
# Print the output.
df
```

```python
# DataFrame using arrays.
import pandas as pd
 
# initialize data of lists.
data = {'Name':['Tom', 'Jack', 'nick', 'juli'],
        'marks':[99, 98, 95, 90]}
 
# Creates pandas DataFrame.
df = pd.DataFrame(data, index =['rank1',
                                'rank2',
                                'rank3',
                                'rank4'])
 
# print the data
df
```

#**Dict形式でデータフレームを作成**

```python
# Python code demonstrate how to create
# Pandas DataFrame by lists of dicts.
import pandas as pd
 
# Initialize data to lists.
data = [{'a': 1, 'b': 2, 'c':3},
        {'a':10, 'b': 20, 'c': 30}]
 
# Creates DataFrame.
df = pd.DataFrame(data)
 
# Print the data
df
```

```python
# Python code demonstrate to create
# Pandas DataFrame by passing lists of
# Dictionaries and row indices.
import pandas as pd
 
# Initialize data of lists
data = [{'b': 2, 'c':3}, {'a': 10, 'b': 20, 'c': 30}]
 
# Creates pandas DataFrame by passing
# Lists of dictionaries and row index.
df = pd.DataFrame(data, index =['first', 'second'])
 
# Print the data
df
```

```python
import pandas as pd
  
# Initialize lists data.
data = [{'a': 1, 'b': 2},
        {'a': 5, 'b': 10, 'c': 20}]
  
# With two column indices, values same
# as dictionary keys
df1 = pd.DataFrame(data, index =['first',
                                 'second'],
                   columns =['a', 'b'])
  
# With two column indices with
# one index with other name
df2 = pd.DataFrame(data, index =['first',
                                 'second'],
                   columns =['a', 'b1'])
  
# print for first data frame
print (df1, "\n")
  
# Print for second DataFrame.
print (df2)
```

#**空のデータフレームを作成して書き込んでいくパターン**

#空のデータフレームを作成

```python
# カラムがないindexだけ設定されているDataFrameを作成
df = pd.DataFrame(index=[], columns=[])
print(df)
```

```python
#列を書き込んでいく
df["age"] = [10, 15, 20, 25]
df["sex"] = ["M", "F", "M", "F"]
df
```

```python
#行数を揃えないとエラーになる
df["name"] = ["Yamada", "Sato", "Tanaka"]
```

#先に列名だけ決めておき行を追加していくパターン

```python
# カラムがないindexだけ設定されているDataFrameを作成
df = pd.DataFrame(index=[], columns=["age","sex","name"])
print(df)
```

```python
df.loc[len(df.index)] = [10, "M", "Yamada"] 
df.loc[len(df.index)] = [15, "F", "Sato"] 
df
```

```python
#列数が揃っていないとエラーになる
df.loc[len(df.index)] = [15, "F"] 
```
